```python
import pandas
import time, datetime
import numpy as np
from pandas import DataFrame
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.cross_validation import KFold
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
from IPython.display import display
from sklearn.preprocessing import scale
from sklearn.linear_model import LogisticRegression
```

```python
features = pandas.read_csv('/home/ubuntu/Desktop/Coursera ML/week7/final project/features.csv')
features = features.iloc[:, 2:]

display(features.head(5))
```

```python
NA_features = {'Features': [], 'NAs, %': []}
filled_values = dict(features.count())
for feature in filled_values.keys():
    if filled_values[feature] != len(features):
        NA_features['Features'].append(feature)
        NA_features['NAs, %'].append((len(features) - filled_values[feature]) / float(len(features)))
NA_features = DataFrame(NA_features)

display(NA_features)
```

```python
print min(features['dire_courier_time'] / float(60))
print max(features['dire_courier_time'] / float(60))
```

```python
features.fillna(value=999,inplace=True)
```

```python
cols_del = ['duration',
            'radiant_win',
            'tower_status_radiant',
            'tower_status_dire',
            'barracks_status_radiant', 
            'barracks_status_dire']
data = features.drop(cols_del, axis=1)
labels = features['radiant_win']
```

```python
display(data.head(5))

data = data.astype(float)
for cols in list(data.columns.values):
    data[cols] = scale(data[cols], copy=False)

display(data.head(5))
```

```python
ROC_time = {'ROC_AUC': [], 'Time': [], 'n_estimators': [10, 20, 30]}
for estimators in [10, 20, 30]:
    start_time = datetime.datetime.now()

    result = 0
    kf = KFold(len(labels), n_folds=5, random_state=1, shuffle=True)
    for train_index, test_index in kf:
        X_train, X_test = np.array(data.iloc[train_index, :]), np.array(data.iloc[test_index, :])
        y_train, y_test = np.array(labels[train_index]), np.array(labels[test_index])   
        GB = GradientBoostingClassifier(n_estimators=estimators, verbose=True, random_state=1)
        GB.fit(X_train, y_train)
        y_pred = GB.predict_proba(X_test)[: ,1]
        result += roc_auc_score(y_test, y_pred)
    result = result / 5.     

    ROC_time['ROC_AUC'].append(result)
    ROC_time['Time'].append(datetime.datetime.now() - start_time)
    #print result    
    #print 'Time elapsed:', datetime.datetime.now() - start_time
ROC_time = DataFrame(ROC_time)
    
display(ROC_time)
```

```python
features = pandas.read_csv('/home/ubuntu/Desktop/Coursera ML/week7/final project/features.csv')
features = features.iloc[:, 2:]

NA_features = {'Features': [], 'NAs, %': []}
filled_values = dict(features.count())
for feature in filled_values.keys():
    if filled_values[feature] != len(features):
        NA_features['Features'].append(feature)
        NA_features['NAs, %'].append((len(features) - filled_values[feature]) / float(len(features)))
NA_features = DataFrame(NA_features)

features.fillna(value=0,inplace=True)

cols_del = ['duration',
            'radiant_win',
            'tower_status_radiant',
            'tower_status_dire',
            'barracks_status_radiant', 
            'barracks_status_dire']
data = features.drop(cols_del, axis=1)
labels = features['radiant_win']



from sklearn.preprocessing import StandardScaler
data = data.astype(float)
data = StandardScaler().fit_transform(data)
display(data)

ROC_time = {'ROC_AUC': [], 'Time': [], 'C': [0.0001, 0.001, 0.1, 1, 10, 100, 10000]}
for reg in [0.0001, 0.001, 0.1, 1, 10, 100, 10000]:
    start_time = datetime.datetime.now()
    
    result = 0
    kf = KFold(len(labels), n_folds=5, random_state=1, shuffle=True)
    for train_index, test_index in kf:
        X_train, X_test = np.array(data[train_index, :]), np.array(data[test_index, :])
        y_train, y_test = np.array(labels[train_index]), np.array(labels[test_index])   
        logist = LogisticRegression(C=reg, n_jobs=-1, random_state=1)
        logist.fit(X_train, y_train)
        y_pred = logist.predict_proba(X_test)[: ,1]
        result += roc_auc_score(y_test, y_pred)
    result = result / 5.     
    
    print result
    
    ROC_time['ROC_AUC'].append(result)
    ROC_time['Time'].append(datetime.datetime.now() - start_time)
    #print result    
    #print 'Time elapsed:', datetime.datetime.now() - start_time
ROC_time = DataFrame(ROC_time)
    
display(ROC_time)
```
